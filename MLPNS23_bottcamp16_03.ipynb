```python
import pylab as pl
import numpy as np
import scipy as sp
from scipy import stats
import seaborn
seaborn.set_style("darkgrid")
```

```python
np.random.seed(456)
mymean = 100
```

```python
df = mymean #sto usando la distribuzione chi quadro quindi i gdl li setto al valore medio
p=0.5 #se distribuzione binomiale
n= mymean/p

mydist = {} #creo un sample per la mia distribuzione

pl.hist(np.random.chisquare(mymean, size=500));
pl.ylabel('N')
pl.xlabel('x');
```

```python
mysize = (np.random.rand(100) * 2000).astype(int)
mysize[mysize <10] = 10 #setto i valori minori di 10 a 10
pl.plot(mysize, '.');
```

```python
for s in mysize:
  mydist[s] = np.random.chisquare(mymean, size=s)

ax = pl.figure().add_subplot(111)

for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.')
ax.plot(pl.xlim(), [100, 100], 'k', alpha=0.5)
ax.set_xlabel("sample size")
ax.set_xlabel("sample mean");
```
